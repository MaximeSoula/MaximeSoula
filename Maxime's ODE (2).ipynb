```python
import numpy as np 
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from numpy import exp

V_0 = -70
T = 10 #Si on a T = 2 ou T = 100, on aura des fluctuations très étranges
lamda = 0.5

def ODE(V_0, N, T):
    delta = T/N
    time = np.arange(0, T, delta)
    mylist = []
    V = V_0
    for i in time :
        V =  V - lamda*delta*V
        mylist.append(V)
    print(len(mylist))
    return mylist, time
    
    
a, time = ODE(V_0, 50, T) #faible mesure si on ne fait que sur 10 points (N= 10) 
a2, time2 = ODE(V_0, 100, T)
a3, time3 = ODE(V_0, 1000, T)


plt.figure(figsize= (15,5))
plt.xlabel("temps en ms")
plt.ylabel("Potentiel de membrane")
plt.plot(time, a)
plt.plot(time2, a2)
plt.plot(time3, a3)



def solODE(V_0, N, T):
    state = []
    delta = T/N
    time = np.arange(0, T, delta)
    for t in time:
        Vt = V_0*exp(-lamda*t)
        state.append(Vt)
    print(len(state))
    return state, time


u, time = solODE(V_0, 50, T) 
plt.plot(time, u)
```
