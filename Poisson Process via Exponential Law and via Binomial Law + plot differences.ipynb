```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sps


n = 600
T = 60
dt = T/n

t = np.linspace(0, T, n)

def Poisson_process_bin():
    global n, T, dt, t
    time = 0
    spikelist = []
    totalevents = []
    while time < T:
        time = time +dt
        Xvariable = np.random.binomial(1, 0.7*dt)
        spikelist.append(Xvariable)
        if Xvariable == 1:
            totalevents.append(time)
    print(totalevents)
    print(spikelist)
    plt.plot(spikelist,'o')
    plt.show()
    plt.plot(np.arange(0, len(totalevents)), totalevents, 'o')
    
a = Poisson_process_bin()
        
        
```
